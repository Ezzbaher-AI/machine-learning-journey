```python
import pandas as pd

data = {
    "Hours_Studied": [2, 3, 4, 5, 6, 7, 8, 9],
    "Sleep_Hours": [5, 6, 6, 7, 8, 8, 9, 10],
    "Exam_Score": [50, 60, 65, 75, 90, 85, 80, 70],
    "Attendance": [60, 80, 90, 50, 95, 50, 70, 65]
}

df = pd.DataFrame(data) #df is DataFrame

#print(df.head())

x = df[["Hours_Studied", "Sleep_Hours", "Attendance"]]
y = df["Exam_Score"]

print("Features:")
print(x.shape)

print("\nTarget")
print(y.shape)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Attendance"], df["Exam_Score"])

plt.xlabel("Attendance")
plt.ylabel("Exam Score")
plt.title("Attendance vs Exam Score")

plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)

print("Training Features Shape:", x_train.shape)
print("Testing Features Shape:", x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train, y_train)

print("Model trained successfully!")

print(model.coef_)
print(model.intercept_)
```

```python
predictions = model.predict(x_test)

print("Predictions:")
print(predictions)

print("\nActual Values:")
print(y_test.values)
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(y_test, predictions)

print("Mean Absolute Error:", error)
```

```python
new_student = pd.DataFrame({
    "Hours_Studied": [7],
    "Sleep_Hours": [8],
    "Attendance": [90]
})

predicted_score = model.predict(new_student)

print(predicted_score)
```

```python
train_score = model.score(x_train, y_train)
test_score = model.score(x_test, y_test)

print("Training Score:", train_score)
print("Testing Score:", test_score)
```
